```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import LogisticRegression, DecisionTreeClassifier, RandomForestClassifier, GBTClassifier, NaiveBayes
```

```python
spark=SparkSession.builder.appName('Classification').getOrCreate()
```

```python
data=[(25,50000,0),(35,65000,1),(45,80000,1),(23,40000,0),(52,90000,1)]
df=spark.createDataFrame(data,['age','salary','label'])
```

```python
dataset=VectorAssembler(inputCols=['age','salary'],outputCol='features').transform(df).select('features','label')
train,test=dataset.randomSplit([0.8,0.2],42)
```

```python
LogisticRegression(labelCol='label').fit(train).transform(test).show()
```

```python
DecisionTreeClassifier(labelCol='label').fit(train).transform(test).show()
```

```python
RandomForestClassifier(labelCol='label').fit(train).transform(test).show()
```

```python
GBTClassifier(labelCol='label').fit(train).transform(test).show()
```

```python
NaiveBayes(labelCol='label').fit(train).transform(test).show()
```
