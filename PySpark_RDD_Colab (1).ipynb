***RDD is the core data structure of Spark.***

**RDD (Resilient Distributed Dataset)** is the fundamental data structure in Apache Spark.

It stores data in a distributed manner across multiple machines, allowing Spark to process the data in parallel.

RDD is the way Spark organizes data for processing.

Spark divides the data into smaller partitions.
Each partition is processed by a different machine simultaneously.
This collection of partitions is called an RDD

RDD

R -- Resilient (Recovers from failures)
-- Spark automatically recreates the missing partition using its execution history.

D -- Distributed (Spread across machines)
-- The data is distributed across multiple machines instead of being stored on a single machine.
Each machine processes its own partition.

D -- Dataset (Collection of records)

builder is used to configure the Spark application before creating it.

# SparkContext is the core engine of Spark.
Before DataFrames and Spark SQL were introduced, SparkContext was the main entry point to Spark.

It is responsible for:

Connecting to the Spark cluster

Distributing tasks

Creating RDDs

Managing executors

Because RDDs are created using SparkContext

SparkSession is the modern, high-level interface used for DataFrames, SQL, and ML.


SparkContext is the lower-level engine that manages the Spark cluster and RDDs.

```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('RDD').getOrCreate()
sc=spark.sparkContext
```

```python
rdd=sc.parallelize([1,2,3,4,5])
rdd.collect()
```

parallelize() Convert a Python collection (like a list) into an RDD so Spark can process it in parallel.

collect() retrieves all the elements of the RDD and brings them back to the Driver program as a Python list.

```python
rdd.map(lambda x:x*x).collect()
```

```python
rdd.filter(lambda x:x%2==0).collect()
```

```python
sc.parallelize(['hello world','hello pyspark']).flatMap(lambda x:x.split()).map(lambda x:(x,1)).reduceByKey(lambda a,b:a+b).collect()
```

However, writing code using RDDs requires more programming effort.
